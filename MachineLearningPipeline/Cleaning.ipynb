```python
# =============================================================================
# TYPE CONVERSION FUNCTION
# =============================================================================

def convert_data_types(df, dtype_dict, date_columns):
    """
    Convert DataFrame columns to appropriate data types.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        The input DataFrame
    dtype_dict : dict
        Dictionary mapping column names to desired data types
    date_columns : dict
        Dictionary mapping date column names to their format strings
        
    Returns:
    --------
    pandas.DataFrame
        DataFrame with converted data types
    """
    import pandas as pd
    from datetime import datetime
    
    df_converted = df.copy()
    conversion_log = []
    
    print("Starting data type conversions...")
    print("=" * 50)
    
    # Convert basic data types
    for col, dtype in dtype_dict.items():
        if col in df_converted.columns:
            try:
                original_dtype = str(df_converted[col].dtype)
                
                # Special handling for boolean columns
                if dtype == "boolean":
                    # Convert 0/1 to True/False, handle NaN
                    df_converted[col] = df_converted[col].astype("boolean")
                
                # Special handling for categorical columns
                elif dtype == "category":
                    df_converted[col] = df_converted[col].astype("category")
                
                # Standard numeric conversions
                else:
                    df_converted[col] = df_converted[col].astype(dtype)
                
                conversion_log.append(f"✓ {col}: {original_dtype} → {dtype}")
                
            except Exception as e:
                conversion_log.append(f"✗ {col}: Failed to convert - {str(e)}")
    
    # Convert date columns
    for col, date_format in date_columns.items():
        if col in df_converted.columns:
            try:
                original_dtype = str(df_converted[col].dtype)
                
                if col == "pl_pubdate":
                    # Special handling for year-month format
                    df_converted[col] = pd.to_datetime(df_converted[col], format=date_format, errors='coerce')
                else:
                    # Standard date parsing
                    df_converted[col] = pd.to_datetime(df_converted[col], format=date_format, errors='coerce')
                
                conversion_log.append(f"✓ {col}: {original_dtype} → datetime64[ns] (format: {date_format})")
                
            except Exception as e:
                conversion_log.append(f"✗ {col}: Failed to convert date - {str(e)}")
    
    # Print conversion summary
    successful = len([log for log in conversion_log if log.startswith("✓")])
    failed = len([log for log in conversion_log if log.startswith("✗")])
    
    print(f"Conversion Summary:")
    print(f"✓ Successful: {successful}")
    print(f"✗ Failed: {failed}")
    print("=" * 50)
    
    # Print detailed log
    for log in conversion_log:
        print(log)
    
    return df_converted, conversion_log

print("Type conversion function created successfully!")
```

```python
# HELPER FUNCTIONS FOR COLUMN INFORMATION
# =============================================================================

def get_column_info(column_name):
    """
    Get detailed information about a column.
    
    Parameters:
    -----------
    column_name : str
        The new (renamed) column name
        
    Returns:
    --------
    dict or None
        Dictionary with old_name and description, or None if not found
    """
    return column_info.get(column_name, None)

def search_columns_by_keyword(keyword):
    """
    Search for columns containing a keyword in their name or description.
    
    Parameters:
    -----------
    keyword : str
        Keyword to search for (case-insensitive)
        
    Returns:
    --------
    list
        List of matching column names
    """
    keyword = keyword.lower()
    matches = []
    
    for col_name, info in column_info.items():
        if (keyword in col_name.lower() or 
            keyword in info['description'].lower() or 
            keyword in info['old_name'].lower()):
            matches.append(col_name)
    
    return matches

def display_column_info(column_name):
    """
    Pretty print column information.
    
    Parameters:
    -----------
    column_name : str
        The new (renamed) column name
    """
    info = get_column_info(column_name)
    if info:
        print(f"Column: {column_name}")
        print(f"Original name: {info['old_name']}")
        print(f"Description: {info['description']}")
    else:
        print(f"Column '{column_name}' not found in column_info dictionary")

def display_columns_by_category(category_list, category_name):
    """
    Display information for all columns in a category.
    
    Parameters:
    -----------
    category_list : list
        List of column names
    category_name : str
        Name of the category for display
    """
    print(f"\n{category_name.upper()} COLUMNS")
    print("=" * 50)
    
    for col in category_list:
        if col in column_info:
            print(f"• {col}")
            print(f"  Original: {column_info[col]['old_name']}")
            print(f"  Description: {column_info[col]['description']}")
            print()

print("Helper functions created:")
print("• get_column_info(column_name)")
print("• search_columns_by_keyword(keyword)")
print("• display_column_info(column_name)")
print("• display_columns_by_category(category_list, category_name)")
```

### Notes on Expanded Hyperparameter Searches

The parameter grids were expanded for both XGBoost and Decision Tree models:

**XGBoost** additions:
- Added higher `n_estimators` range (300–800)
- Added broader `max_depth` values (4, 6, 10)
- Added stronger regularization knobs: `gamma` (0,1,2) and `reg_lambda` (1,2)
- Kept `learning_rate`, `subsample`, and `colsample_bytree` ranges moderate to balance exploration vs runtime.

Estimated combinations: 648 (before CV). With 5-fold CV that implies 3240 model fits.

**Decision Tree** additions:
- Included `splitter`, `max_features`, and pruning parameter `ccp_alpha`.
- Dynamic trimming logic reduces the grid if raw combination count exceeds 4000 to avoid excessive runtime.

If runtime becomes prohibitive consider:
1. Switching to `RandomizedSearchCV` with `n_iter=100`.
2. Using a smaller CV (e.g., 3-fold) during exploration, then refit with 5-fold.
3. Narrowing ranges around currently best-performing regions.
4. Early feature selection or dimensionality reduction (e.g., PCA, mutual information).

You can also log intermediate results by converting `grid_search.cv_results_` / `grid_dt.cv_results_` into a DataFrame and sorting by mean test score for further analysis.
